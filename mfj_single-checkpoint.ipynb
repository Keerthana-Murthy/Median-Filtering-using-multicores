```julia

using Images
img = load("1st_noise.jpg")
pixels1 = (permutedims(channelview(img), (2,3,1)))
println(size(pixels1))
@time begin
function median_filter(pixels,filter_size)
    
    mid1 = filter_size / 2
    mid = trunc(Int,mid1)
    temp = []
    c = 1
    for l in 1:size(pixels)[1]-filter_size
        for k in 1:size(pixels)[1]-filter_size
            for i in l:(l+filter_size-1)
                d=c+i
                p=d-l
                for j in k:(k+filter_size-1)
                    append!(temp,pixels[i,j,1])
                if p == filter_size
                    u = p+j
                if (p%filter_size == 0) && ((u-(k-1))%filter_size == 0)
                    sort!(temp)
                    
                    mid2=trunc(Int,size(temp)[1]/2)
                    pixels[i-mid,j-mid,1]=temp[mid2+1]
                    deleteat!(temp, 1:(size(temp)[1]))
                    end   
                        
                
                
                        end
                end
                
            end
        end
    end
    for l in 1:size(pixels)[1]-filter_size
        for k in 1:size(pixels)[1]-filter_size
            for i in l:(l+filter_size-1)
                d=c+i
                p=d-l
                for j in k:(k+filter_size-1)
                    append!(temp,pixels[i,j,2])
                if p == filter_size
                    u = p+j
                if (p%filter_size == 0) && ((u-(k-1))%filter_size == 0)
                    sort!(temp)
                    
                    mid2=trunc(Int,size(temp)[1]/2)
                    pixels[i-mid,j-mid,2]=temp[mid2+1]
                    deleteat!(temp, 1:(size(temp)[1]))
                    end   
                        
                
                
                        end
                end
                
            end
        end
    end
    for l in 1:size(pixels)[1]-filter_size
        for k in 1:size(pixels)[1]-filter_size
            for i in l:(l+filter_size-1)
                d=c+i
                p=d-l
                for j in k:(k+filter_size-1)
                    append!(temp,pixels[i,j,3])
                if p == filter_size
                    u = p+j
                if (p%filter_size == 0) && ((u-(k-1))%filter_size == 0)
                    sort!(temp)
                    
                    mid2=trunc(Int,size(temp)[1]/2)
                    pixels[i-mid,j-mid,3]=temp[mid2+1]
                    deleteat!(temp, 1:(size(temp)[1]))
                    end   
                        
                
                
                        end
                end
                
            end
        end
    end
    R = pixels

    imgc2 = colorview(RGB, permuteddimsview(R, (3,1,2)))
    
end

median_filter(pixels1,5)
end
```
